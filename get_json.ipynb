```python
import requests
from requests.structures import CaseInsensitiveDict

# Define the URL of the exoplanet page
url = "https://www.exoplanetkyoto.org/exohtml/Kepler-758_d.html"

# Define the headers with proper formatting
headers = CaseInsensitiveDict()
headers["Accept"] = "text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8"
headers["Accept-Language"] = "en-GB,en-US;q=0.9,en;q=0.8"
headers["User-Agent"] = (
    "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
    "AppleWebKit/537.36 (KHTML, like Gecko) "
    "Chrome/124.0.0.0 Safari/537.36"
)

# Optionally, you can add other headers if necessary
# headers["Referer"] = "https://www.google.com/"
# headers["Connection"] = "keep-alive"

try:
    # Send the GET request
    response = requests.get(url, headers=headers, timeout=10)

    # Check if the request was successful
    response.raise_for_status()

    # Print the status code
    print(f"Status Code: {response.status_code}")

    # Optionally, print a snippet of the content
    print("Content Snippet:")
    print(response.text)  # Print first 500 characters

except requests.exceptions.HTTPError as http_err:
    print(f"HTTP error occurred: {http_err}")  # e.g., 404 Not Found
except requests.exceptions.ConnectionError as conn_err:
    print(f"Connection error occurred: {conn_err}")  # e.g., DNS failure, refused connection
except requests.exceptions.Timeout as timeout_err:
    print(f"Timeout error occurred: {timeout_err}")  # e.g., request timed out
except requests.exceptions.RequestException as req_err:
    print(f"An error occurred: {req_err}")  # Any other request-related errors
```

```python
import json
from bs4 import BeautifulSoup
import re
import logging

logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')


def parse_html(html_content):
    soup = BeautifulSoup(html_content, 'html.parser')
    data = {
        
        "Images": [],
        "Descriptive Lists": []
    }

    # Parse images
    for img in soup.find_all('img'):
        data["Images"].append({
            "Description": img.get('alt', ''),
            "Source": img['src']
        })

    # Parse descriptive lists
    for li in soup.find_all('li'):
        if li.b:
            data["Descriptive Lists"].append(li.b.text.strip())


    return data

def main():
    try:
        with open('input.html', 'r', encoding='utf-8') as f:
            html_content = f.read()
        
        data = parse_html(html_content)
        
        with open('kepler_758_b_data.json', 'w', encoding='utf-8') as f:
            json.dump(data, f, indent=4, ensure_ascii=False)
        
        logging.info("JSON file 'kepler_758_b_data.json' has been successfully created.")
    except FileNotFoundError:
        logging.error("Input HTML file not found. Please ensure 'input.html' exists in the current directory.")
    except Exception as e:
        logging.error(f"An error occurred: {str(e)}")

if __name__ == "__main__":
    main()
```
